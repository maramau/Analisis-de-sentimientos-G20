Adapto los datos de los csv al formato establecido por Fernando
date	t	s1	s2	s3	s4	s5	s6	s7	topic	country	year

```python
import pandas as pd
```

```python
dicTopics = {0:"Economia", 1:"Medio ambiente", 2:"Sociedad", 3:"Gobierno"}

dicCountries = {'en': ['alemania', 'arabia', 'australia', 'brasil', 'canada', 'china', 'corea del sur', 'francia', 
                        'india', 'indonesia', 'italia', 'japon', 'uk', 'rusia', 'sudafrica', 'turquia', 'eu', 'usa'],
                'es': ['argentina', 'mexico'],
                'pt':['brasil']}
cantTopics = 4
years = [2018, 2020, 2021]
langs = ["es", "en"]
```

```python
dicDf = {"date":[], "t":[], "No emo":[], "Joy":[], "Anger":[], "Sadness":[],
           "Disgust":[], "Fear":[], "Trust":[], "Surprise":[], "Anticipation":[],
           "topic":[], "country":[], "year":[]}
jumps = 0

for topic in range(cantTopics):
    for year in years:
        for lang in langs:
            folder = "predictions\\Lexicon\\" + str(year) + "\\" + lang + "\\"
            for country in dicCountries[lang]:
                i = 1
                filename = country + " " + str(year) + " - Tema " + str(topic) + ".csv"
                try:
                    df = pd.read_csv(folder + filename, encoding="utf-8", sep="\t")
                    print("Read file: " + filename)
                    jumps = 0
                    for j in range(len(df)):
                        dicDf["date"].append("/".join(df["date"][j].split("-")))
                        dicDf["t"].append(i)
                        dicDf["No emo"].append(df["No emotion"][j])
                        dicDf["Joy"].append(df["Joy"][j])
                        dicDf["Anger"].append(df["Anger"][j])
                        dicDf["Sadness"].append(df["Sadness"][j])
                        dicDf["Disgust"].append(df["Disgust"][j])
                        dicDf["Fear"].append(df["Fear"][j])
                        dicDf["Trust"].append(df["Trust"][j])
                        dicDf["Surprise"].append(df["Surprise"][j])
                        dicDf["Anticipation"].append(df["Anticipation"][j])
                        dicDf["topic"].append(str(dicTopics[topic]))
                        dicDf["country"].append(country)
                        dicDf["year"].append(str(year))
                        i = i+1
                except FileNotFoundError:
                    print("Missing file: " + filename)
                if (jumps == 0):
                    dicDf["date"].append("")
                    dicDf["t"].append("")
                    dicDf["No emo"].append("")
                    dicDf["Joy"].append("")
                    dicDf["Anger"].append("")
                    dicDf["Sadness"].append("")
                    dicDf["Disgust"].append("")
                    dicDf["Fear"].append("")
                    dicDf["Trust"].append("")
                    dicDf["Surprise"].append("")
                    dicDf["Anticipation"].append("")
                    dicDf["topic"].append("")
                    dicDf["country"].append("")
                    dicDf["year"].append("")
                    jumps += 1
                    
#dfNew = new DataFrame(headers)
```

```python
df = pd.DataFrame(dicDf)
```
