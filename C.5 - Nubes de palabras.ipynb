La idea de este archivo es desarrollar codigo que, dado el contenido de los tweets procesados, cuente las apariciones de palabras en ellos y produzca nubes de palabras

```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
model = 'Plutchick'
dic_lang_countries = {'en':['alemania', 'arabia', 'brasil', 'australia', 'canada', 'china', 'corea del sur', 'francia', 
                            'india', 'indonesia', 'italia', 'japon', 'uk', 'rusia', 'sudafrica', 'turquia', 'eu', 'usa'],
                      'es':['argentina', 'mexico'],
                      'pt':['brasil']}
years = [2018, 2020, 2021]
langs = ['es', 'en', 'pt']
cant_temas = 4
```

```python
cant_temas = 4
for year in years:
    for lang in langs:
        for country in dic_lang_countries[lang]:
            for subject in range(cant_temas):
                print(str(year) + " " + lang +" " + country + " " + str(subject))
                path = 'results_processed/'+str(year)+'/'+lang+'/'
                file = country + ' ' + str(year) + ' - Tema ' + str(subject) + '.csv'
                df = pd.read_csv(path + file, sep = '\t')
                text = " ".join(tweet for tweet in df.text.astype('U'))
                if (len(text) > 10):
                    mask = np.array(Image.open('img/flags/'+country+'.png'))
                    wordcloud = WordCloud(background_color="white", mode="RGB", max_words=1000, mask=mask, 
                                          stopwords=None, normalize_plurals=True, repeat=False, include_numbers=False, 
                                          min_word_length=2, min_font_size=5, collocations=True).generate(text)

                    # create coloring from image
                    image_colors = ImageColorGenerator(mask)
                    plt.figure(figsize=[12,9])
                    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
                    plt.axis("off")

                    # store to file
                    new_path = 'reports/wordclouds/' + model + '/' + str(year) + '/' + lang + '/'
                    new_file = 'wordcloud - ' + str(year) + ' ' + country + ' Tema ' + str(subject) + '.png'
                    plt.savefig(new_path + new_file, format="png")
                    plt.close()
                    #plt.show()
```
